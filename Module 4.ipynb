```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
# Load CSV normally
df = pd.read_csv("vgsales.csv")

# Select only numeric columns for clustering
numeric_df = df[['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales', 'Global_Sales']]

# Drop rows with any missing values (if there are any)
numeric_df = numeric_df.dropna()
```

```python
# Inertia values for elbow method
inertia_values = []
for k in range(2, 31):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(numeric_df)
    inertia_values.append(kmeans.inertia_)

# Plotting elbow graph (optional)
plt.plot(range(2, 31), inertia_values, marker='o')
plt.xlabel("Number of Clusters")
plt.ylabel("Inertia")
plt.title("Elbow Method for Optimal k")
plt.show()

print(inertia_values)
```

```python
prev = inertia_values[0]
for inertia in inertia_values[1:]:
    print(prev-inertia)
    prev=inertia
```
